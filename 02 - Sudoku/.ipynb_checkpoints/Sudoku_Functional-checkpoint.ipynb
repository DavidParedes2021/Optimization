```python
import matplotlib.pyplot as plt
import numpy   as np
import pandas  as pd
import seaborn as sns
```

```python
test_sudoku = np.loadtxt(open("sudokus/sudoku1.csv"), delimiter=',', dtype=int)

def get_missing_idx(matrix):
	missing_idx = []
	for line in matrix:
		line_missing = []
		for i, val in enumerate(line):
			if val == 0: line_missing.append(i)
		missing_idx.append(line_missing)
	return missing_idx

missing_idx = get_missing_idx(test_sudoku)

def print_sudoku(sudoku, missing_idx=None):
	for i, line in enumerate(sudoku):
		if i%3 == 0: print('\033[0m')
		for j, val in enumerate(line):
			if j%3 == 0: print('\033[0m|', end='')
			val_str = None
			head = '\033[0m'

			if val > 0:
				val_str = str(val)
			else:
				val_str = '-'

			if missing_idx is not None:
				if j in missing_idx[i]:
					head = '\033[1;32m'
			else:
				if val == 0:
					head = '\033[1;31m'
			
			print(head+val_str, end=' ')
		#-----------------
		print('\033[0m')
	#-----------------
[print(line) for line in missing_idx]
print_sudoku(test_sudoku, missing_idx=None)

def get_max_fitness(matrix):
	return 9*9 - np.count_nonzero(matrix)

max_fitness = get_max_fitness(test_sudoku)
print()
print('Max fitness:', max_fitness)
```

```python
def get_missing_val(matrix):
	missing_val = []
	for line in matrix:
		has     = line[line.nonzero()]
		has_not = []
		for i in range(1, 10):
			if i not in has:
				has_not.append(i)
		missing_val.append(has_not)
	return missing_val

missing_val = get_missing_val(test_sudoku)

for i, genotype in enumerate(missing_val):
	if i%3 == 0: print()
	print(genotype)
print('---------------------------')
for i, missing in enumerate(missing_idx):
	if i%3 == 0: print()
	print(missing)
```

```python
copy_sudoku = test_sudoku.copy()
# print_sudoku(copy_sudoku)

def update_phenotype(phenotype, missing_idx, genotype_vals):
	for i, line in enumerate(phenotype):
		line[missing_idx[i]] = genotype_vals[i]
	return phenotype
update_phenotype(copy_sudoku, missing_idx, missing_val)

def get_fitness(phenotype, missing_idx=None, return_valid=False):
	fitness = 0
	if return_valid:
		valid = []
	# mi, mj = phenotype.shape
	check = [[j for j in range(9)] for i in range(9)]
	if missing_idx is not None:
		check = missing_idx

	for i, line in enumerate(check):
		for jidx in line:
			val = phenotype[i,jidx]
			i_cuadrant = int(i/3)   *3
			j_cuadrant = int(jidx/3)*3

			i_count = np.count_nonzero(phenotype[i]       == val)
			j_count = np.count_nonzero(phenotype[:, jidx] == val)
			c_count = np.count_nonzero(phenotype[i_cuadrant:i_cuadrant+3, j_cuadrant:j_cuadrant+3] == val)

			if i_count != 1: continue
			if j_count != 1: continue
			if c_count != 1: continue
			
			fitness += 1
			if return_valid: valid.append( (i,jidx) )
	if return_valid:
		return fitness, valid
	return fitness

test_fitness = get_fitness(copy_sudoku, missing_idx=missing_idx)

# print('Original----------')
# print_sudoku(test_sudoku)
print('Copy with genotype')
print_sudoku(copy_sudoku, missing_idx=missing_idx)

print(test_fitness)
print(get_fitness(copy_sudoku))
# print(copy_sudoku[valid_ones[1]])
```

```python
def mix_parents(genotype_a, genotype_b, mix_ratio:float=0.5):
	genotype_len = len(genotype_a)
	middle: int  = int(genotype_len*mix_ratio)
	child = [genotype_a[idx].copy() for idx in range(0, middle)]+[genotype_b[idx].copy() for idx in range(middle, genotype_len)]
	# child = genotype_a[0:middle].copy() + genotype_b[middle:genotype_len].copy()
	return child

child_genotype = mix_parents(random_genotypes[0], random_genotypes[1])
print(len(random_genotypes[0]))
print(len(child_genotype))
print('child genotype:')
print_genotype(child_genotype)
print()
```

```python
def mutate(genotype, rng_gen=np.random.default_rng, return_changes:bool=False):
	rng = rng_gen()
	genotype_len = len(genotype)
	genotype_cpy = [chromosome.copy() for chromosome in genotype]
	line_mut = int(rng.integers(low=0, high=genotype_len, size=1))

	chromosome = genotype_cpy[line_mut]
	chromosome_len = len(chromosome)

	gene_idx = rng.integers(low=0, high=chromosome_len, size=2)
	# print(chromosome)
	temp = chromosome[gene_idx[0]]
	chromosome[gene_idx[0]] = chromosome[gene_idx[1]]
	chromosome[gene_idx[1]] = temp

	if return_changes:
		return genotype_cpy, (line_mut, gene_idx)

	return genotype_cpy

mutated_genotype, (line_mut, gene_idx) = mutate(random_genotypes[0], return_changes=True)

print(f'mutation in line: {line_mut}')
print(gene_idx)
print('Original')
# print_genotype(random_genotypes[0])
print(random_genotypes[0][line_mut])
print('mutated')
# print_genotype(mutated_genotype)
print(mutated_genotype[line_mut])
```

```python
def genetic_sudoku(
	original_sudoku
	,max_k:int=1000
	,n_P:int=100
	,rng_gen=np.random.default_rng
	,P_cross:float=0.2
	,P_mut:float=0.1
	,MAX_F_LEN:int = 30
	,MAX_K_LEN:int = 20
	,SPINNER = ['\\','|', '/', '-']
):
	rng = rng_gen()
	copy_sudoku = original_sudoku.copy()
	missing_val = get_missing_val(original_sudoku)
	missing_idx = get_missing_idx(original_sudoku)
	max_fitness = get_max_fitness(original_sudoku)

	SPINNER_LEN = len(SPINNER)

	Gi   = [[rng.permutation(line) for line in missing_val] for i in range(n_P)]
	Gip1 = [None for i in range(n_P)]
	Fi = [
		get_fitness(update_phenotype(copy_sudoku, missing_idx, Gi_i), missing_idx=missing_idx)
		for Gi_i in Gi
	]
	elite_idx = np.argmax(Fi)

	elite = Gi[elite_idx]
	Fe    = Fi[elite_idx]

	k=0
	while k < max_k and Fe < max_fitness:
		k += 1
		for i, Genotype in enumerate(Gi):
			if rng.random() < P_cross:
				parent_idx = rng.integers(low=0, high=n_P, size=4)
				parent_a = parent_idx[np.argmax([Fi[idx] for idx in parent_idx[:2]])]
				parent_b = parent_idx[np.argmax([Fi[idx] for idx in parent_idx[2:4]])+2]
				Gip1[i] = mix_parents(Gi[parent_a], Gi[parent_b])
			else:
				Gip1[i] = Genotype.copy()
			
			if rng.random() < P_mut:
				# number_of_mut = int(rng.integers(low=0, high=10, size=1))
				Gip1[i] = mutate(Gi[i])
				# for it_ in range(number_of_mut):
				# 	Gip1[i] = mutate(Gip1[i])
		# ------------------
		Gi = Gip1
		Fi = [
			get_fitness(update_phenotype(copy_sudoku, missing_idx, Gi_i), missing_idx=missing_idx)
			for Gi_i in Gi
		]
		elite_idx = np.argmax(Fi)
		update_progress_bar:bool = False
		if Fe < Fi[elite_idx]:
			elite = Gi[elite_idx]
			Fe    = Fi[elite_idx]
			update_progress_bar = True

		if k%10 == 1 or update_progress_bar:
			filled_k = int(MAX_K_LEN * k  // max_k)
			filled_f = int(MAX_F_LEN * Fe // max_fitness)
			k_bar = '#' * filled_k + '-' * (MAX_K_LEN - filled_k)
			f_bar = '#' * filled_f + '-' * (MAX_F_LEN - filled_f)
			finess_update = ' '
			if update_progress_bar:
				finess_update = '!'

			print(f'\r{SPINNER[int(k/10)%SPINNER_LEN]}', end='')
			print(f' Fitness:[{f_bar}]{finess_update}{max_fitness} |', end='')
			print(f' Iteration:[{k_bar}] {max_k}: {k/float(max_k)*100:03.1f}%', end='\r')
	#---------------
	print(f'\nfitness: {Fe}/{max_fitness} achieved on iteration {k}')
	return update_phenotype(copy_sudoku, missing_idx, elite)

solved = genetic_sudoku(test_sudoku, P_mut=0.4, P_cross=0.6, n_P=80, max_k=50000)
```
